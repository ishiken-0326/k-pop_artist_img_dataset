```python
from bs4 import BeautifulSoup
import urllib
from urllib.request import urlopen

url = 'https://dbkpop.com/db/female-k-pop-idols'

headers = {
'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:47.0) Gecko/20100101 Firefox/47.0'
}

request = urllib.request.Request(url, headers=headers)
html = urlopen(request)
bs_obj = BeautifulSoup(html, 'html.parser')
```

```python
table = bs_obj.find('table', {'id': 'table_1'})

col_name = []
for item in table.findAll('tr')[0].findAll('th')[:-1]:
    text = item.get_text()
    col_name.append(text)

res_list = []
for row in table.findAll('tr')[2:]:
    res_row = []

    items = row.findAll('td')[:-1]
    profile = items[0].find('a')
    if profile is not None:
        profile_link = profile.get('href')
    else:
        profile_link = None
    res_row.append(profile_link)

    for item in items[1:-1]:
        text = item.get_text()
        res_row.append(text)

    instagram = items[-1].find('a')
    if instagram is not None:
        instagram_link = instagram.get('href')
    else:
        instagram_link = None
    res_row.append(instagram_link)

    res_list.append(res_row)
```

```python
import pandas as pd

artist_info_pd = pd.DataFrame(res_list, columns=col_name)
artist_info_pd.to_csv('artist_info_pd_femal.csv', index=False)
```

## download img from google

```python
artist_names = artist_info_pd['Korean Stage Name']
artist_fullnames = artist_info_pd['Full Name']
group_names = artist_info_pd['Group']
keyword_list = []

# for name in zip(artist_names, group_names):
#     if name[1] != '':
#         keyword = ' '.join(name) + ' 셀카'
#         keyword_list.append(keyword)

for name in zip(artist_names, group_names):
    if name[1] != '':
        keyword_list.append(' '.join(name))
```

```python
from google_images_download import google_images_download 
response = google_images_download.googleimagesdownload()

for keyword in keyword_list:
    try:
        arguments = {"keywords":keyword, 
                     "limit":1000,
                     "size": '>800*600',
                     'format': 'png',
                     'chromedriver': '/usr/local/bin/chromedriver',
                     'silent_mode': True,
                     'type': 'face'
                    } 
        response.download(arguments)
        print('Success png DL' + keyword)
    except:
        print('Error png DL : ' + keyword)

    try:
        arguments = {"keywords":keyword,
                     "limit":1000,
                     "size": '>800*600',
                     'format': 'jpg',
                     'chromedriver': '/usr/local/bin/chromedriver',
                     'silent_mode': True,
                     'type': 'face'
                    } 
        response.download(arguments)
        print('Success jpg DL' + keyword)
    except:
        print('Error jpg DL : ' + keyword)
```
